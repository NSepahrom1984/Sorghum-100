```python
import glob
import numpy as np
import cv2

x= []
dicit = {'a' : 1}
counter = 0
dim =(128,128)
files = glob.glob("C:/Users/16395/Desktop/Sorghom/train_images/*.png")

for file in files:
    image = cv2.imread(file)
    print(counter)
    counter = counter + 1
    if counter ==40000:
        break
    image = cv2.resize(image , dim ,interpolation = cv2.INTER_AREA)
    avgg = np.average(image)
    dicit[file] = avgg
    x.append(image)
```
